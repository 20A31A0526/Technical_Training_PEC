### **Set**
1. s1.intersection(s2)
   - gives common elements in both sets
2. s1.union(s2)
   - gives all the elements from both the sets without duplicates
3. s1.difference(s2)
   - returns elements from the s1 which are not present in s2
4. s1.add(element)
   - adds element into the list
5. s1.difference_update(s2)

```python
s1={1,2,3,4}
s2={2,5,6,1}
print('intersection:',s1.intersection(s2))
print(s1.union(s2))
print('difference:',s1.difference(s2))
s1.add(5)
print('after add:',s1)
s1.difference_update(s2)
print('difference update:',s1)
```

### **Conditional Statements**
1. if(condtition):

        statements
2. elif(condition):

        statements

3. else:
  
        statements
Types of 

```python
#even or odd
a=int(input())
print('Using if-else:')
if a%2==0:
  print(a,'is even')
else:
  print(a,'is odd')
print('Using ternary statement:')
print(a,'is even') if a%2==0 else print(a,'is odd')
```

### **Ternary Operator**
- print(statement) if condition else print(statement)

```python
#printing week days
a=int(input('Enter the day number :'))
if a==1:
  print('Sunday')
elif a==2:
  print('Monday')
elif a==3:
  print('Tuesday')
elif a==4:
  print('Wednesday')
elif a==5:
  print('Thursday')
elif a==6:
  print('Friday')
elif a==7:
  print('Saturday')
else:
  print('Please enter valid number')
```

```python
#>0 and <20 if even print weird  else normal
#20-30 print normal 
#>=30 and odd print normal else weird
num=int(input())
if num>0 and num<20:
  if num%2==0:
    print('Weird Number')
  else:
    print('Normal Number')
elif num>=20 and num<30:
  print('Normal Number')
else:
  if num%2==0:
    print('Weird Number')
  else:
    print('Normal Number')
```

### **Range function**
- range(start,end,step_size)


```python
#range function
for i in range(1,10):
  print(i,end=' ')
```

```python
import math as m
l=[(12,63),(25,38),(26,55)]
for i in l:
  a=m.prod(i)
  print(a*a,end=' ')
  
```

```python
l=[10,20,30,40]
n=int(input())
if n not in l:
  print(-1)
  exit(0)
for i in range(len(l)):
  if l[i]==n:
    print('Index is:',i)
    break
```

### **List Comprehension**
- list=[x for x in range(start,end)]

```python
#list comprehension vs normal method
l=[]
for i in range(10):
  l.append(i)
print(l)
#another way
l=[x for x in range(10)]
print(l)
```

```python
#generate even numbers in 0-50 range
l=[num for num in range(0,51) if num%2==0]
print(l)
```

```python
#printing 
l=[i for i in range(1,101) if i%7==0 and i%11==0]
print(l)
```

```python
#printing list in reverse order
l=[1,2,3,4]
for i in range(len(l)-1,-1,-1):
  print(l[i],end=' ')
```

```python
#sum of positive and negative elements
l=list(map(int,input().split()))
p,n=0,0
for i in l:
  if i>-1:
    p=p+i
  else:
    n=n+i
print('Sum of positive numbers:',p)
print('Sum of negative numbers:',n)
```
